```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
sui = cv.imread("brown-eyes.jpg")
sui_copy =sui.copy()
gray = cv.cvtColor(sui_copy,cv.COLOR_BGR2GRAY)

circles = cv.HoughCircles(gray,cv.HOUGH_GRADIENT,dp=1,minDist=90,minRadius=1,maxRadius=70,param1=180,param2=40)
circles = np.uint16(np.around(circles))
print(circles.shape)
print(circles)

for i in circles[0, :]:
        center = (i[0], i[1])
        radius = i[2]
     
        cv.circle(sui_copy, center, radius, (0, 255, 0), 2)


plt.imshow(cv.cvtColor(sui_copy,cv.COLOR_BGR2RGB))
plt.axis("off") 
plt.show()
```
