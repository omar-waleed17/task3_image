```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

def process(img):
   
    gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
    median = cv.medianBlur(gray, 75)  
    canny = cv.Canny(median, 2, 80) 
    return canny

def create_masks(img, contours, min_area):
    
    masked_images = [] 
    for contour in contours:
        area = cv.contourArea(contour)
        if area > min_area: 
           
            mask = np.zeros(img.shape[:2], np.uint8)
            
            
            cv.drawContours(mask, [contour], -1, 255, -1)

           
            masked = cv.bitwise_and(img, img, mask=mask)
            masked_images.append(masked)
    return masked_images


img = cv.imread('Ttpvb.jpg')


img_edges = process(img)
contours, hierarchy = cv.findContours(img_edges, cv.RETR_EXTERNAL, cv.CHAIN_APPROX_SIMPLE)


min_contour_area = 1000 


masked_images = create_masks(img, contours, min_contour_area)


for masked in masked_images:
    
    masked_rgb = cv.cvtColor(masked, cv.COLOR_BGR2RGB)
    plt.imshow(masked_rgb)
    plt.axis('off')  
    plt.show()
```
